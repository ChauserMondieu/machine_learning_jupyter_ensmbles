# Gradient Descend
## 最速梯度下降法


```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline
```

<p>假设原函数为</p>
$$ f(x,y) = (x-10)^2 + (y-10)^2 $$
<p>则原函数的Jacobian矩阵为：</p>
$$
\begin{pmatrix}
\frac{\partial f(x,y)}{\partial x} &
\frac{\partial f(x,y)}{\partial y}
\end{pmatrix} 
= 
\begin{pmatrix}
2(x-10) &
2(y-10)
\end{pmatrix}
$$
<p>那么对于原函数来说，其更新方程为</p>
$$
f(x^{k+1},y^{k+1}) = f(x^{k},y^{k}) - {\alpha} {\nabla} f(x^{k},y^{k})
$$
<p>在以上方程中${\alpha}$代表学习速率，也就是步长</p>

```python
def f(x,y):
    return (x-10)**2 + (y-10)**2

# alpha stands for learning rate
# iter stands for iteration times
# ori_x gives original value for x
# ori_y gives original value for y
def gradient_descend(alpha, iter):
    x = 20
    y = 20
    index = 1
    while(iter):
        x -= alpha * 2*(x-10)
        y -= alpha * 2*(y-10)
        print("========================")
        print("%d th iteration, x=%f, y=%f, final value=%f" % (index,x,y,f(x,y)) )
        index += 1
        iter -= 1
    
    
if __name__ == "__main__":
    gradient_descend(0.05,100)
```

<p>加载 $matplotlib$ 中的 $Axes3D$ 模块生成3D画板</p>
<p>使用 $meshgrid$ 操作生成网格，并使用 $plot\_surface$ 方法绘制3D平面</p>
<p>最后将梯度下降法用图像表示出来</p>

```python
def prepare(alpha, iter):
    fig = plt.figure()
    #ax = plt.axes(projection='3d')
    ax = fig.add_subplot(111,projection='3d')
    x = np.linspace(0,20,100)
    y = np.linspace(0,20,100)
    nx,ny = np.meshgrid(x,y)
    nz = f(nx,ny)
    ax.plot_surface(nx, ny, nz, rstride=1, cstride=1, cmap=plt.get_cmap('rainbow'))
    x = 20
    y = 20
    while(iter):
        xb = x
        yb = y
        zb = f(x,y)
        x -= alpha * 2*(x-10)
        y -= alpha * 2*(y-10)
        z = f(x,y)
        ax.scatter3D(x, y, z, 'ko', lw=0.5, ls='-')
        iter -= 1
    plt.show()
```

```python
prepare(0.01,100)
```

## 机器学习中的梯度下降法
<p>机器学习中的梯度下降法主要应用在损失函数 $loss\_function$ 中</p>
<p>对于一般的LR算法，损失函数 $loss\_function$ 为</p>
$$
J({\theta}) = \frac{1}{2N}\{{\sum}^{N}_{i=1}[y_i-(\theta_i x_i + b_i)]^2\}
$$
<p>则损失函数 $J(\theta)$ 对于 $\theta$ 的偏导数为</p>
$$
\frac{\partial J(\theta)}{\partial \theta_i} = \frac{1}{N}{\sum}^{N}_{i=1}x_i(\theta_i x_i + b_i)
$$
<p>*注意，以上公式中的 $i$ 实际上是数据标签而不是每个数据中的维度标签，我们实际上假设所有数据只有一个维度</p>
