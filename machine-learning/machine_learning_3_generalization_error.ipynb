# 机器学习模型的评价
## 偏差 & 方差 & 误差
机器学习本质上就是通过在traning数据集上训练以便最好把握模型数据集的特征，然后在testing数据集上运行模型得到想要的结果 <br>
因此机器学习模型在training数据集上loss以及在testing数据集上loss的差异就是所谓的误差【Generalizating Error】

真正的误差是由于模型本身的偏差【bias】+方差【variance】共同作用产生的：
$$
error = bias + variance
$$
<ul>
偏差表现在模型本身的拟合能力，偏差过大可能说明我们本身对于模型做出了错误的假设；
</ul>
<ul>
方差表现在模型的泛化性能，方差过大可能说明模型的参数过多，存在过拟合的情况；
</ul>

误差的计算公式需要使用到所有训练集合 $Training set$ 的训练结果平均值：
$$
bias = (\frac{1}{N}{\sum}_{d=1}^{N}X_d - Y)
$$
其中 $d$ 代表一个训练集<br>
而方差的计算公式同样需要用到训练集结果的平均值
$$
f(\overline x) = \frac{1}{N}{\sum}_{d=1}^{N}X_d
$$
$$
variance(X) = E[(X-f(\overline x))^2]
$$
